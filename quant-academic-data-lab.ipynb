```python
import pandas as pd
import numpy as np
import math
#df = pd.DataFrame(np.random.randint(0,100, size = (1000, 12)), columns = list('ABCDEFGHIJKL'))
df=(pd.read_csv('d_synth.csv'))
#UPDATE this cell with your data source + Dictionary/Codebook
#data source: https://www.census.gov/data/datasets/time-series/demo/cps/cps-asec.2021.html
#data dictionary: https://www2.census.gov/programs-surveys/cps/datasets/2021/march/asec2021_ddl_pub_full.pdf
#import necessary libaries
import matplotlib
#name file path (you will have download your own dataset and populate the path and choose the appropriate cell to execute) # <-- edit this value! 
#you will run just ONE of the next four code blocks, depending on the file type of your dataset
#read data from CSV (.csv) & print head
#read data from SAS (.sas7bdat or .sas) & print head
#read data from excel (.xls or .xlsx) & print head
#read data from Stata (.dta) & print head
#list columns in data frame
#output frequency table of chosen column in dataset
#prints updated frequency table
col = 'prop_vaccinated' #<-edit this value! column name as it appears in dataframe
col_name = 'data number' #<-edit this value! descriptive name for column
freq = df[col].rename_axis(col_name).reset_index(name='% of vaccinated')
print(freq)
freq = freq.sort_values(by=col_name)
print(freq)
updated_list=freq['% of vaccinated'].values.tolist()
print(updated_list)
#create frequency chart
#list datatype 
def calculations(num_list): 
    #for loop
    sum=0
    for x in num_list:
        sum=sum+x
    n=float(len(num_list))
    mean=sum/n
    print(mean)
    difference=0
    #intialize an empty list
    delta_mean_list= []
    for i in num_list:
            d_mean=i-mean
            delta_mean_list.append(d_mean)
    print(delta_mean_list)
    delta_mean_squared_list=[]
    for d_mean in delta_mean_list:
        d_mean_sqrd=d_mean**2
        delta_mean_squared_list.append(d_mean_sqrd)
    print(delta_mean_squared_list)
    delta_mean_squared_sum=0
    for j in delta_mean_squared_list:
        delta_mean_squared_sum=j+delta_mean_squared_sum
    print(delta_mean_squared_sum)
    variance=delta_mean_squared_sum/n
    standard_deviation=math.sqrt(variance)
    median=np.median(updated_list)
    return mean, standard_deviation, median
print(calculations(updated_list))
```
